```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_ascii_frequency(string):
    ascii_values = [ord(char) for char in string]
    ascii_counts = Counter(ascii_values)

    sorted_ascii_values = sorted(ascii_counts.keys())
    frequencies = [ascii_counts[val] for val in sorted_ascii_values]
    plt.figure(figsize=[20,5])
    plt.bar(sorted_ascii_values, frequencies)
    plt.xlabel('Kode ASCII')
    plt.ylabel('Frekuensi')
    plt.title('Plot Kode ASCII dan Frekuensi')
    plt.xticks(sorted_ascii_values, [str(val) for val in sorted_ascii_values], rotation='vertical')
    # plt.grid(True)
    plt.show()

# Contoh penggunaan
string = "Hello, world!y "
plot_ascii_frequency(string)
string2 = "world hella"
plot_ascii_frequency(string2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def histogram_lines_graph(text1, text2, title):
    ascii_values1 = [ord(char) for char in text1]
    ascii_values2 = [ord(char) for char in text2]

    ascii_counts1 = Counter(ascii_values1)
    ascii_counts2 = Counter(ascii_values2)

    keys = range(256)
    values = range(len(keys))

    plt.plot(values, [ascii_counts1.get(key, 0) for key in keys], 'b-', label='Text 1')
    plt.plot(values, [ascii_counts2.get(key, 0) for key in keys], 'r-', label='Text 2')

    plt.title(title)
    plt.xlabel("Kode ASCII")
    plt.ylabel("Frekuensi")
    # plt.xticks(values, keys)
    plt.legend()
    plt.grid(True)
    plt.show()

# Contoh penggunaan
text1 = "Hello, world!y"
text2 = "laj9ii2---1"
title = 'Histogram Lines Graph'
plt.figure(figsize=[20,5])
histogram_lines_graph(text1, text2, title)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def plot_ascii_frequency(string, title):
    ascii_values = [ord(char) for char in string]
    ascii_counts = Counter(ascii_values)

    y_values = np.arange(256)
    frequencies = [ascii_counts[val] for val in y_values]

    sorted_indices = np.argsort(y_values)
    sorted_frequencies = [frequencies[i] for i in sorted_indices]

    plt.bar(y_values, sorted_frequencies)
    plt.xlabel('Kode ASCII')
    plt.ylabel('Frekuensi')
    plt.title(title)
    plt.show()

with open('ciphertext_trivium/ciphertext_ascii3.txt', 'r', encoding="utf-8") as file:
    string = file.read()
plot_ascii_frequency(string, "test")
```
